```python
import os
from PIL import Image
```

```python
# ----- Setup ----- 
idun = False

def analyze_dataset(is_iphone:bool):
    data_root = "/datasets/tdt4265/Poles2025/"

    if is_iphone:
        dataset = "Road_poles_iPhone" 
        training_set_path   = os.path.join(data_root, dataset, "images", "Train", "train")
        validation_set_path   = os.path.join(data_root, dataset, "images", "Validation", "val")
        test_set_path   = os.path.join(data_root, dataset, "images", "Test", "test")
    else:
        dataset = "roadpoles_v1"
        training_set_path   = os.path.join(data_root, dataset, "train", "images")
        validation_set_path   = os.path.join(data_root, dataset, "valid", "images")
        test_set_path   = os.path.join(data_root, dataset, "test", "images")


    print(f"----- {dataset} -----")
    print(f"  --- Size ---")
    # Get amount of images
    print(f"We have {len(os.listdir(training_set_path))} training images")
    print(f"We have {len(os.listdir(validation_set_path))} validation images")
    print(f"We have {len(os.listdir(test_set_path))} test images")

    # Get resolution of first image (All have same res)

    print(f"\n  --- Resoluton ---")
    files = os.listdir(training_set_path)
    first_img = os.path.join(training_set_path, files[0])
    with Image.open(first_img) as img:
        w, h = img.size
        print(f"{w}x{h}")
```

```python
analyze_dataset(False)
```

```python
analyze_dataset(True)
```
